## Arithmetic Operators
Arithmetic operators are used to perform common mathematical operations in Python. These operators work with numeric data types such as integers and floats.

### Basic Arithmetic Operators

| Operator | Description         | Example        | Result |
|----------|---------------------|----------------|--------|

## Logical Operators
Logical operators are used to combine multiple conditions (Boolean expressions) and return `True` or `False` depending on the logic applied.

### Basic Logical Operators

| Operator | Description                  | Example                  | Result      |
|----------|------------------------------|--------------------------|-------------|
| `and`    | Logical AND (both true)      | `True and False`         | `False`     |
| `or`     | Logical OR (at least one)    | `True or False`          | `True`      |
| `not`    | Logical NOT (negation)       | `not True`               | `False`     |

---

### Truth Table

### `and` Operator
| A     | B     | A and B |
|-------|-------|----------|
| True  | True  | True     |
| True  | False | False    |
| False | True  | False    |
| False | False | False    |

### `or` Operator
| A     | B     | A or B  |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | True    |
| False | True  | True    |
| False | False | False   |

### `not` Operator
| A     | not A |
|-------|-------|
| True  | False |
| False | True  |

---

# Loops

Loops are used to execute a block of code repeatedly. Python supports two main types of loops: `for` loops and `while` loops.

---

### 1. `for` Loop

A `for` loop is used to iterate over a sequence such as a list, tuple, string, or range.

#### Syntax

A variable takes each value in the sequence one at a time, and the loop executes once for each value.
```python
for variable in sequence: 
    # code block
```
---

### 2. `while` Loop

A `while` loop continues to execute as long as a specified condition is `True`.

#### Syntax
The condition is evaluated before each iteration. If the condition becomes `False`, the loop stops.

---

## Loop Control Statements

Python includes special statements that change the normal flow of loops:

| Statement | Description                             |
|-----------|-----------------------------------------|
| `break`   | Exits the loop immediately              |
| `continue`| Skips the current iteration             |
| `pass`    | Does nothing; used as a placeholder     |

---

## `else` with Loops

Both `for` and `while` loops can have an optional `else` block.

- The `else` block runs **only if the loop completes without a `break`**.
- If the loop ends due to a `break` statement, the `else` block is skipped.

---

## List Comprehension

List comprehension provides a concise way to create lists using a single line of code. It is more readable and often faster than traditional loops.

It combines `for` loops and optional conditional logic into a single expression.

- Can be nested or used with functions for transformations.

List comprehensions are ideal when:
- You want to create a new list from an existing iterable.
- The transformation logic is simple and readable in one line.

## Writing Functions in Python

Functions are reusable blocks of code that perform a specific task. They help organize code, avoid repetition, and improve readability.

### Defining a Function

In Python, functions are defined using the `def` keyword, followed by the function name, parentheses with optional parameters, and a colon. The code block inside the function is indented.

#### Example:
```python
def greet():
    print("Hello, world!")
```

````python
def greet():
    print("Hello, world!")

greet()
````

#### Parameters and Arguments
- Parameters are variables listed in the function definition.
- Arguments are values passed into the function when it is called.

````python
def greet(name="Guest"):
    print("Hello,", name)

greet()         # Uses default
greet("Alice")  # Overrides default
````

### Return Statement
Functions can return values using the return statement.

````python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
````

### Variable-Length Arguments
Use `*args` for any number of positional arguments, and `**kwargs` for keyword arguments.

````python
def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(1, 2, 3, 4)
````

````python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

print_info(name="Alice", age=25)
````

### Scope of Variables
- Local variables exist only within a function.
- Global variables are accessible throughout the program.

````python
x = 10  # global

def show():
    x = 5  # local
    print("Inside function:", x)

show()
print("Outside function:", x)
````

If `x` is defined outside all functions (i.e., at the `global` level), and you want to update it from inside a function, use the global keyword:

````python
x = 10 # in the global environment

def use_global():
    
    global x
    return x

use_global()
````

If `x` is defined in an outer function (`not global`), and you want to update it in an inner (nested) function, use the `nonlocal` keyword.

````python
def outer():
    x = 10  # enclosing scope

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()
````

# Object-Oriented Programming

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of **objects**, which can contain data (**attributes**) and code (**methods**). Python supports OOP and makes it easy to create and use classes and objects.

---

## Key Concepts

| Concept      | Description                                                                 |
|--------------|-----------------------------------------------------------------------------|
| Class        | A blueprint for creating objects                                             |
| Object       | An instance of a class                                                       |
| Attribute    | A variable that belongs to a class or object                                 |
| Method       | A function that belongs to a class                                           |
| `__init__`   | The constructor method called when an object is created                      |
| `self`       | A reference to the current instance of the class                             |
| Inheritance  | A way to create a new class using properties and methods of an existing one  |

| Polymorphism | Ability to redefine methods for different types of objects                   |

## Saving and  Loading data
NumPy can:

* Store data in its own fast, compact binary format (.npy) with np.save() and reload it with `np.load()`.
* Save arrays as text files like CSV using `np.savetxt()` for human-readable or spreadsheet use.
* Load text data (e.g., CSV) back into arrays with `np.genfromtxt()`.


As you can see, the data is presented in a wide format and the columns names `F_TL` (Total number of Female) has two information which are Gender and Value. Also, `M_20_24` means count of male in the age category` 20-24`.

The task is to reshape the data from wide to long format. For that, the pandas `melt` could be used as follows.

````python
cat_vars = ['year', 'ISO3', 'ADM0_FR', 'ADM0_PCODE', 'ADM1_FR', 'ADM1_PCODE',
       'ADM2_FR', 'ADM2_PCODE']
rwpop_melted = pd.melt(rwpop, id_vars=cat_vars, value_name="Population")
rwpop_melted.head()
````

Now, let's filter out rows containing totals `T_`, `_TL`

````python

rwpop_melted['variable'][30]
indx = np.array([x[0] in ["F", "M"] and "TL" not in x.split("_") for x in rwpop_melted['variable']]) # type: ignore
rwpop_melted = rwpop_melted[indx]
rwpop_melted
````

Bingo! We've removed all the totals from the data. Now, the next cleaning is to separate the column variable into `variable` in to `gender` and `age_group`. After that, decode `F` to `Female` and `M` to `Male`.

````python
# TODO: Separate variable to gender and age_group
rwpop_melted['age_group'] = [x[2:] for x in rwpop_melted.variable]
# rwpop_melted['Gender'] = [x.split("_")[0] for x in rwpop_melted.variable]
rwpop_melted['Gender'] = [x[:1] for x in rwpop_melted.variable]

# TODO: decode the variable gender
rwpop_melted.loc[:, 'Gender'] = rwpop_melted['Gender'].replace({'F':'Female', 'M':'Male'})
````
